# Import

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import pandas as pd
```

# Set up driver

```python
options = webdriver.ChromeOptions()
options.add_argument("--disable-Advertisement")
options.add_argument("--disable-popup-blocking")
capabilities = options.to_capabilities()
capabilities = {
    "browser": "chrome",
    "browser_version": "latest",
    "os": "Windows",
    "os_version": "11",
    "build": "Python Sample Build",
    "name": "Pop-ups testing",
}

driver = webdriver.Chrome(options=options)
```

# Login and load full page

```python
driver.get("https://github.com/login")
wait = WebDriverWait(driver, 3)
wait.until(EC.element_to_be_clickable((By.NAME, "login"))).send_keys(
    "mathonline03@gmail.com"
)
wait.until(EC.element_to_be_clickable((By.ID, "password"))).send_keys("fithcmus2003")
wait.until(EC.element_to_be_clickable((By.NAME, "commit"))).click()

driver.get("https://accounts.topdev.vn/login/github")
driver.get("https://topdev.vn/viec-lam-it")

num_record = int(
    driver.find_element(
        By.CSS_SELECTOR, "h1.text-xl.font-bold.text-black span.text-primary"
    ).text
)

driver.find_element(
    By.CSS_SELECTOR, "button.tw-cursor-pointer.tw-border-0.tw-rounded-md"
).click()

driver.find_element(
    By.XPATH,
    "//button[contains(@class, 'font-bold') and contains(@class, 'uppercase') and contains(text(), 'Vi')]",
).click()

driver.maximize_window()

current_num_record = 0
temp = 0

timeout = 60  # Maximum wait time (seconds) when no change occurs
start_time = time.time()  # Capture the start time

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollTop);")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    temp = len(driver.find_elements(By.CSS_SELECTOR, ".mb-4.last\\:mb-0"))

    if current_num_record < temp:
        current_num_record = temp
        print(current_num_record)
        start_time = time.time()  # Reset start time if there is a change

    if current_num_record >= num_record:
        break

    # Check if the time has exceeded the timeout with no changes
    if time.time() - start_time > timeout:
        print("Timeout: No change detected within the specified time.")
        break
```

# Get all job links

```python
driver.find_elements(By.CSS_SELECTOR, ".mb-4.last\\:mb-0")[0].find_element(
    By.TAG_NAME, "a"
).get_attribute("href")

url_list = []
for ele in driver.find_elements(By.CSS_SELECTOR, ".mb-4.last\\:mb-0"):
    if ele.find_elements(By.TAG_NAME, "a"):
        url_list.append(ele.find_element(By.TAG_NAME, "a").get_attribute("href"))
url_list
```

# Crawl

```python
data = []

display(len(url_list))

for i in range(len(url_list)):
    driver.get(url_list[i])

    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located(
            (By.CSS_SELECTOR, "h1.text-2xl.font-bold.text-black")
        )
    )

    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located(
            (By.XPATH, '//h3[contains(text(),"Ngnh ngh")]/following-sibling::p')
        )
    )

    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located(
            (
                By.XPATH,
                '//h3[contains(text(),"Nm kinh nghim ti thiu")]/following-sibling::div/a',
            )
        )
    )

    ten_cong_viec_elements = driver.find_elements(
        By.CSS_SELECTOR, "h1.text-2xl.font-bold.text-black"
    )
    ten_cong_viec = [element.text for element in ten_cong_viec_elements]

    ten_cong_ty_elements = driver.find_elements(
        By.CSS_SELECTOR, "p.text-base.font-bold.text-gray-500"
    )
    ten_cong_ty = [element.text for element in ten_cong_ty_elements]

    muc_luong_elements = driver.find_elements(
        By.XPATH, "//div[contains(@class, 'flex max-w-[540px] items-center')]/p"
    )
    muc_luong = [element.text for element in muc_luong_elements]

    dia_chi_elements = driver.find_elements(
        By.XPATH, "//div[contains(@class, 'flex last:mb-0')]/div"
    )
    dia_chi = [element.text for element in dia_chi_elements][:-1]

    ngay_dang_elements = driver.find_elements(
        By.XPATH, '//div[contains(text(), "ng")]'
    )
    ngay_dang = [[element.text for element in ngay_dang_elements][-1]]

    nganh_nghe_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Ngnh ngh")]/following-sibling::p'
    )
    nganh_nghe = [element.text for element in nganh_nghe_elements]

    quy_mo_cong_ty_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Quy m cng ty")]/following-sibling::p'
    )
    quy_mo_cong_ty = [element.text for element in quy_mo_cong_ty_elements]

    quoc_tich_cong_ty_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Quc tch cng ty")]/following-sibling::p'
    )
    quoc_tich_cong_ty = [element.text for element in quoc_tich_cong_ty_elements]

    nam_kinh_nghiem_elements = driver.find_elements(
        By.XPATH,
        '//h3[contains(text(),"Nm kinh nghim ti thiu")]/following-sibling::div/a',
    )
    nam_kinh_nghiem = [element.text for element in nam_kinh_nghiem_elements]

    cap_bac_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Cp bc")]/following-sibling::div'
    )
    cap_bac = [element.text for element in cap_bac_elements]

    loai_hinh_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Loi hnh")]/following-sibling::div/a'
    )
    loai_hinh = [element.text for element in loai_hinh_elements]

    loai_hop_dong_elements = driver.find_elements(
        By.XPATH, '//h3[contains(text(),"Loi hp ng")]/following-sibling::div/a'
    )
    loai_hop_dong = [element.text for element in loai_hop_dong_elements]

    cong_nghe_su_dung_elements = driver.find_elements(
        By.XPATH,
        '//h3[contains(text(),"Cc cng ngh s dng")]/following-sibling::div//a/span',
    )
    cong_nghe_su_dung = [element.text for element in cong_nghe_su_dung_elements]

    quy_trinh_phong_van_elements = driver.find_elements(
        By.XPATH,
        '//h3[contains(text(),"Quy trnh phng vn")]/following-sibling::div//li',
    )
    quy_trinh_phong_van = [element.text for element in quy_trinh_phong_van_elements]

    record = [
        ten_cong_viec,
        ten_cong_ty,
        muc_luong,
        dia_chi,
        ngay_dang,
        nganh_nghe,
        quy_mo_cong_ty,
        quoc_tich_cong_ty,
        nam_kinh_nghiem,
        cap_bac,
        loai_hinh,
        loai_hop_dong,
        cong_nghe_su_dung,
        quy_trinh_phong_van,
        url_list[i],
    ]

    data.append(record)

    print(i)
```

# Save data

```python
columns = [
    "ten_cong_viec",  # Job title
    "ten_cong_ty",  # Company name
    "muc_luong",  # Salary
    "dia_chi",  # Address
    "ngay_dang",  # Posting date
    "nganh_nghe",  # Industry
    "quy_mo_cong_ty",  # Company size
    "quoc_tich_cong_ty",  # Company nationality
    "nam_kinh_nghiem",  # Years of experience
    "cap_bac",  # Level
    "loai_hinh",  # Type
    "loai_hop_dong",  # Contract type
    "cong_nghe_su_dung",  # Technology used
    "quy_trinh_phong_van",  # Interview process
    "url",  # Link to job
]

df = pd.DataFrame(data, columns=columns)
df
```
